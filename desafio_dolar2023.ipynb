```python
#Importação de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt #biblioteca para visualização dos dados
import numpy as np
```

```python
#Leitura dos dados
df_dolar_2023 = pd.read_csv('https://raw.githubusercontent.com/programacaodinamica/dicas-pandas/refs/heads/master/dados/bcdata-dolar-2023.csv', sep=';', decimal=',', encoding='utf-8')
```

```python
#Conversão de tipo da coluna data
df_dolar_2023['data'] = pd.to_datetime(df_dolar_2023['data'], format='%d/%m/%Y')
df_dolar_2023.set_index('data', inplace=True)
```
